```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import random
```

```python
paths = glob.glob("/hdd/sacson_language_rand_15_25/*", recursive=True)
stop = False
```

```python
 
path = random.choice(paths)
print(path)
data_path = os.path.join(path, "traj_data.pkl")
data = np.load(data_path, allow_pickle=True)
image_paths = sorted(glob.glob(os.path.join(path, "*.jpg")), key=lambda x: int(x.split("/")[-1].split(".")[0]))
print("LANGUAGE: ", data["language_annotations"])
fig, ax = plt.subplots(1, len(image_paths), figsize=(40, 20))
for i, image_path in enumerate(image_paths):
    image = plt.imread(image_path)
    ax[i].imshow(image)
    ax[i].axis("off")
plt.show()
```
